```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('DO_Well.csv')
df.head()
```

```
df1=df[['M2','T2','W2','TH2','F2']]
```

```
import numpy as np
```

```
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df_mean=imp_mean.fit_transform(df1)
df_mean=pd.DataFrame(df_mean)
```

```
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
df_median=imp_median.fit_transform(df1)
df_median=pd.DataFrame(df_median)
```

```
from sklearn.impute import KNNImputer
imp_KNN = KNNImputer(n_neighbors=5,weights='uniform')
df_KNN=imp_KNN.fit_transform(df1)
df_KNN=pd.DataFrame(df_KNN)
```

```
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp_mice = IterativeImputer(max_iter=10)
df_mice=imp_mice.fit_transform(df1)
df_mice=pd.DataFrame(df_mice)
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
df['M2'].plot(kind='kde', ax=ax, color='blue',label='Original')
df_KNN[0].plot(kind='kde', ax=ax, color='red',title='KNN',label='KNN')
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
df['M2'].plot(kind='kde', ax=ax, color='blue',label='Original')
df_mean[0].plot(kind='kde', ax=ax, color='green',title='Mean',label='Mean')
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
df['M2'].plot(kind='kde', ax=ax, color='blue')
df_median[0].plot(kind='kde', ax=ax, color='cyan',title='Median',label='Median')
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
df['M2'].plot(kind='kde', ax=ax, color='blue')
df_mice[0].plot(kind='kde', ax=ax, color='pink',title='MICE')
```

```
df_KNN_r2=df_KNN[[0,	1,	2,	3,	4,	'M2',	'T2',	'W2',	'TH2'	,'F2']]
```

```
df_MICE_r2=df_MICE[[0,	1,	2,	3,	4,	'M2',	'T2',	'W2',	'TH2'	,'F2']]
```

```
from sklearn.metrics import r2_score
r2_score(df_KNN_r2[[0,	1,	2,	3,	4]], df_KNN_r2[['M2',	'T2',	'W2',	'TH2'	,'F2']])
```

```
from sklearn.metrics import r2_score
r2_score(df_MICE_r2[[0,	1,	2,	3,	4]], df_MICE_r2[['M2',	'T2',	'W2',	'TH2'	,'F2']])
```
